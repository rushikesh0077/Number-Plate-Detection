#### Importing libraries

```python
import ultralytics
from ultralytics import YOLO
import cv2
import pytesseract
import warnings
warnings.filterwarnings('ignore')
```

#### defining function for detection with pretrained weights

```python
def number_plate_detection(img):
    model = YOLO('numberplate_weights.pt')
    res = model.predict(img)
    img_copy = img.copy()
    if len(res[0].boxes.data) > 0:
        for i in range(len(res[0].boxes.data)):
            if float(res[0].boxes.data[i][4]) > 0.65:
                x1 = int(res[0].boxes.data[i][0])
                y1 = int(res[0].boxes.data[i][1])
                x2 = int(res[0].boxes.data[i][2])
                y2 = int(res[0].boxes.data[i][3])
                img_new = img_copy[y1:y2,x1:x2]
                img_new = cv2.cvtColor(img_new, cv2.COLOR_BGR2RGB)
                pytesseract.pytesseract.tesseract_cmd =r'C:\Users\keshk\AppData\Local\Tesseract-OCR\tesseract.exe'
                text = pytesseract.image_to_string(img_new, lang='eng', config='--psm 10 --oem 1 -c tessedit_char_whitelist=ABCDEFGHIJKLMNOPQRSTUVWXYZ0123456789')
                text = text.replace('\n',' ')
                cv2.rectangle(img_copy,(x1,y1),(x2,y2),(255,0,0),2)
                cv2.putText(img_copy,text,(x1,y1-5),cv2.FONT_HERSHEY_DUPLEX,1,(255,255,255),2)
            else:
                return img
        return img_copy
    else:
        return img
```

#### testing on single image

```python
img = cv2.imread('test_1.jpg')
```

```python
result = number_plate_detection(img)
```

```python
while True:
    cv2.imshow('img',result)

    if cv2.waitKey(3) & 0xff == 27:
        break
cv2.destroyAllWindows()
```

#### Testing on video and realtime camera

```python
cap = cv2.VideoCapture(0)
cap.set(cv2.CAP_PROP_FPS, 60)
while True:
    ret , frame = cap.read()
    img_new = number_plate_detection(frame)
    
    cv2.imshow('frame',img_new)

    k = cv2.waitKey(1)
    if k == 27:
        break
cap.release()
cv2.destroyAllWindows()
```
